```python
import geopandas as gpd

import altair as alt


import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = gpd.read_file('NYC_crime - NYC_crime.csv')
df = gpd.read_file(gpd.datasets.get_path('nybb'))
df.set_index('BoroCode',inplace = True)
df.sort_index(inplace = True)
data.head()
```

```python
gdf = gpd.GeoDataFrame(data,geometry = gpd.points_from_xy(data.Longitude,data.Latitude))
gdf_1 = gdf
```

```python
fig,ax = plt.subplots()
df.to_crs("EPSG:4326").plot(ax=ax,color = "white",edgecolor = "black")
gdf.plot(
    column = 'OFNS_DESC',
    categorical = True,
    legend = True,
    ax = ax, 
)
```

```python
gdf.crs
gdf
```

```python
gdf['OFNS_DESC'].unique()
```

```python
occ = gdf["ADDR_PCT_CD"]
kind = gdf["OFNS_DESC"]
sum1 = 0
sum2 = 0
sum3 = 0
sum4 = 0
sum5 = 0
for i in range(len(occ)):
    if kind[i] == 'SEX CRIMES':
        sum1 += int(occ[i])
    elif kind[i] == 'GRAND LARCENY':
        sum2 += int(occ[i])
    elif kind[i] == 'FRAUDS':
        sum3 += int(occ[i])
    elif kind[i] == 'HARRASSMENT':
        sum4 += int(occ[i])
    elif kind[i] == 'OTHERS':
        sum5 += int(occ[i])

occurance = []
occurance.append(str(sum1))
occurance.append(str(sum2))
occurance.append(str(sum3))

occurance.append(str(sum4))

occurance.append(str(sum5))
kind1 = ['SEX CRIMES','GRAND LARCENY','FRAUDS','HARRASSMENT','OTHERS']
print (occurance)
print(kind1)
```

```python
source = pd.DataFrame({'kind': kind1,'occurance':occurance})
scale = alt.Scale(domain = ['SEX CRIMES','GRAND LARCENY','FRAUDS','HARRASSMENT','OTHERS'],
                range = ['#1f77b4','#ff7f0e','#2ca02c','#9467bd','#17becf'])
color1 = alt.Color('kind:N',scale = scale)
source['kind']

```

```python
brush = alt.selection_interval(encodings=['x'])
click = alt.selection_multi(encodings=['color'])
legend = alt.selection_multi(fields=['kind'],bind = 'legend')

alt.Chart(source).mark_bar().encode(
    alt.X('occurance:Q', title = 'occurance'),
    alt.Y('kind',title = 'Kind'),
    color = alt.condition(click,color1,alt.value('lightgray')),
    opacity = alt.condition(legend,alt.value(1),alt.value(0.2))
).properties(width = 600).add_selection(click,legend)
```

```python
total_2021 = dataframe["year_2021"]
sum_2021 = 0
for i in total_2021:
    sum_2021+=i

total_2001 = dataframe["year_2001"]
sum_2001 = 0
for i in total_2001:
    sum_2001+=i
total_1998 = dataframe["year_1998"]
sum_1998 = 0
for i in total_1998:
    sum_1998+=i
total_1993 = dataframe["year_1993"]
sum_1993 = 0
for i in total_1993:
    sum_1993+=i
total_1990 = dataframe["year_1990"]
sum_1990 = 0
for i in total_1990:
    sum_1990+=i
ans = []
ans.append(sum_1990)
ans.append(sum_1993)
ans.append(sum_1998)
ans.append(sum_2001)
ans.append(sum_2021)
ans
year = ["1990","1993","1998","2001","2021"]

```

```python
source1 = pd.DataFrame({'sum1': ans,'year':year})
```

```python
source1
```

```python
scale1 = alt.Scale(domain = ['1990','1993','1998','2001','2021'],
                range = ['#1f77b4','#ff7f0e','#2ca02c','#9467bd','#17becf'])
bars = alt.Chart(source1).mark_bar().encode(
    alt.Y( "year:T", title = "year"),
    alt.X( "sum1:Q", title = " total cases"),
    color = alt.Color("year:T", scale = scale1)
)
text = bars.mark_text(
    align='left',
    baseline = 'middle',
    dx = 3
).encode(
text = 'sum1:Q')
(bars+text).properties(width = 600)
```

```python
years = pd.read_csv("years_of_crime.csv")
```

```python
brush = alt.selection_interval(encodings=['x'])
click = alt.selection_multi(encodings=['color'])
legend = alt.selection_multi(fields=['kind'],bind = 'legend')

graph = alt.Chart(years).mark_circle(size=60).encode(
    alt.X('year:Q',scale = alt.Scale(domain = [1990,2022])),
    y='cases',
    color = alt.condition(brush,'kind',alt.value('lightgrey')),
    opacity = alt.condition(legend,alt.value(1),alt.value(0.2))
).properties(
    width = 550,
    height = 400
    
).add_selection(
    brush,
    legend   
)
```
